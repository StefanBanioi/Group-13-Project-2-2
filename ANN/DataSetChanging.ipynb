```python
import pandas as pd
data = pd.read_csv('ANN\original_kaggle_healthinsurance.csv')

data.head()
```

```python
data = pd.read_csv('ANN\original_kaggle_healthinsurance.csv')
#drop hereditary_disease column
data_wo_disease = data.drop('hereditary_diseases', axis=1) 
#check if data contains missing values or nan
print(data_wo_disease.isnull().sum())

#drop rows with missing values
data_wo_disease = data_wo_disease.dropna()

print(data_wo_disease.isnull().sum())



# Now apply get_dummies
data_wo_disease = pd.get_dummies(data_wo_disease)

data_wo_disease.replace(False, 0, inplace=True)
data_wo_disease.replace(True, 1, inplace=True)

#put claim column to the end
def move_column_to_end(data, col):
    data[col] = data.pop(col)

# Usage
move_column_to_end(data_wo_disease, 'claim')

#transform pandas back into csv
data_wo_disease.to_csv('without_disease.csv', index=False)

data_wo_disease.head()

```
