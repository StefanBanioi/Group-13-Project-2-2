```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Stefan_dataset.csv')
#Stefan data set has the label encoded job column but city is not label encoded

# Do label encoding for the city column
from sklearn.preprocessing import LabelEncoder

#delete all rows with missing values
df = df.dropna()

le = LabelEncoder()
df['city'] = le.fit_transform(df['city'])



#print number of features 
print("Number of features: ", len(df.columns) - 1)
df.head()
```

```python
# Split into X and y
X = df.drop(columns=['claim'])
y = df['claim']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=69)

# Standardize all columns
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)




```

```python
hidden_layers = 8
neurons = 10

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(neurons, input_shape=(X_train_scaled.shape[1],)))
for _ in range(hidden_layers - 1):
    model.add(tf.keras.layers.Dense(neurons, kernel_regularizer=tf.keras.regularizers.l2(0.01), bias_regularizer=tf.keras.regularizers.l2(0.01), activity_regularizer=tf.keras.regularizers.l2(0.01)))
    model.add(tf.keras.layers.LeakyReLU())
model.add(tf.keras.layers.Dense(1, activation='linear'))

model.compile(optimizer='adamW', loss='mean_squared_error', metrics=['mae', tf.keras.metrics.RootMeanSquaredError(), tf.keras.metrics.MeanAbsolutePercentageError()])

model.fit(X_train_scaled, y_train, epochs=150, batch_size=32)

score_train = model.evaluate(X_train_scaled, y_train)
score_test = model.evaluate(X_test_scaled, y_test)

print("Train Score: ", score_train)
print("Test Score: ", score_test)
```
